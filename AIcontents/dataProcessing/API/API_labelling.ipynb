```python
import pandas as pd

def score_labelling(data,book):
  data=data.loc[:,['sentence','score']]
  data['label']=None
  for i in range(len(data)):
      if (data.iloc[i,1]>=0) and (data.iloc[i,1]<0.15):
          data.iloc[i,2]=1
      elif (data.iloc[i,1]>=0.15):
          data.iloc[i,2]=2
      else:
          data.iloc[i,2]=0
  data=data.dropna(how='any')
  data.to_csv('./' + book + '_labelled.csv', header=True, index=True)
  display(data)

def main():
    books=['Pinocchio','Junglebook']
    for book in books:
      data=pd.read_csv('./'+book+'_API.csv')
      #display(data)
      score_labelling(data,book)
    return 0

if __name__ == "__main__":
	main()
```
