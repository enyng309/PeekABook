```python
import pandas as pd
file=pd.read_csv('./mnt/MyDrive/hand_labelled/Alice-in-Wonderland.csv')
file.set_axis(['sentence','emo'], 
                    axis='columns', inplace=True)
data=file
data.sort_values(by=['emo'], axis=0,inplace=True)
data.reset_index(inplace=True,drop=True)
display(file)
```

```python
emoName=['중립','행복','분노','사랑','슬픔','두려움','호의','놀람','걱정']
emoSize=[]
for j in range(9): #감정개수
    count=0
    for i in range(len(data)):
        if data.iloc[i,1]==j:
            count+=1
    print(emoName[j],'의 갯수:',count)
    emoSize.append(count)
    if j==0:nuetral_size=count
emo_sorted_Size=sorted(emoSize,reverse=True)
print(emo_sorted_Size)
```

```python
#중립갯수 분노*4개로 맞추기
#dataframe 중립 문장들 중에 랜덤으로 뽑아서
#새로 저장하기

import random
import numpy as np

def get_random_Nuetral(data,nuetral_size):
    size=emo_sorted_Size[1]*4
    #nuetral_data=pd.DataFrame(columns=['sentence','emo'])
    np.random.seed(0)
    random_list=np.random.randint(nuetral_size,size=size)
    #print(random_list)
    for i in range(size):
        f = open('./Alice-in-Wonderland_data/0/'+str(i)+".txt", 'w')
        f.write(str(data.iloc[random_list[i],0]))
        f.close()
    
    
    
if emo_sorted_Size[1]*4<emoSize[0]:
  get_random_Nuetral(data,nuetral_size)
```

```python
import nlpaug.augmenter.word as naw
def get_text(data,emo):
    for i in range(len(data)):
        if data.iloc[i,1]==emo:
            aug = naw.ContextualWordEmbsAug(
            model_path='bert-base-uncased', action="insert")
            text=str(data.iloc[i,0])
            
            f = open('./Alice-in-Wonderland_data/'+str(emo)+'/'+str(i)+".txt", 'w')
            f.write(text)
            f.close()
            
            for j in range(3):
                augmented_text = aug.augment(text)
                f = open('./Alice-in-Wonderland_data/'+str(emo)+'/'+str(i)+str(j)+".txt", 'w')
#               print("Original:")
#               print(text)
#               print("Augmented Text:")
#               print(augumented_text)
                f.write(augmented_text)
                f.close()
    

def main():
    for emo in range(1,9):
        get_text(data,emo)
    return 0

if __name__=='__main__':
    main()
```
